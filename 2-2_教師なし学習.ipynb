```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# ■次元削減
- PCA(主成分分析)
- SVD(特異値分解)
- t-SNE
- UMAP

```python
df_skill = pd.read_csv("2-2_skill_level.csv",index_col=0)
```

## 次元削減の実行

### PCA(主成分分析)

```python
from sklearn.decomposition import PCA

model_dim = PCA(n_components=2)
vecs_list = model_dim.fit_transform(df_skill)
```

### SVD(特異値分解)

```python
from sklearn.decomposition import TruncatedSVD

model_dim = TruncatedSVD(n_components=2)
vecs_list = model_dim.fit_transform(df_skill)
```

### t-SNE

```python
from sklearn.manifold import TSNE

model_dim = TSNE(n_components=2, perplexity=3)
vecs_list = model_dim.fit_transform(df_skill)
```

### UMAP

```python
import umap.umap_ as umap
from scipy.sparse.csgraph import connected_components

model_dim = umap.UMAP(n_components=2, n_neighbors=4)
vecs_list = model_dim.fit_transform(df_skill)
```

## 寄与率の計算　※PCA/SVDのみ

```python
model_dim.explained_variance_ratio_
```

## 結果の2次元表示

```python
# x,yの分離
X = vecs_list[:,0]
Y = vecs_list[:,1]

# グラフの作成
sns.set(font="Hiragino Maru Gothic Pro",context="talk")
plt.figure(figsize=(8, 8))
sns.scatterplot(x=X,y=Y)

for i,(x_name,y_name) in enumerate(zip(X,Y)):
    plt.annotate(df_skill.index[i],(x_name,y_name))
plt.show()
```

## Y軸の意味の推察 ※PCA/SVDのみ

```python
X_comp, Y_comp = model_dim.components_

sns.set(font="Hiragino Maru Gothic Pro",context="talk")
plt.figure(figsize=(8, 8))

sns.scatterplot(x=X_comp,y=Y_comp)

for i,(annot_x,annot_y) in enumerate(zip(X_comp,Y_comp)):
    plt.annotate(df_skill.columns[i],(annot_x,annot_y))
plt.show()
```

### K-means

```python
from sklearn.cluster import KMeans

vec = KMeans(n_clusters = 3, n_init='auto')
group_num = vec.fit_predict(df_skill)
```

```python
#クラスタ別集計
df_skill_2 = df_skill.copy()
df_skill_2["グループ名"] = group_num
```

### クラスタリング結果の確認　（クラスタリング×次元削減）

```python
model_dim = TruncatedSVD(n_components=2)
vecs_list = model_dim.fit_transform(df_skill)

X = vecs_list[:,0]
Y = vecs_list[:,1]

sns.set(font="Hiragino Maru Gothic Pro",context="talk")
plt.figure(figsize=(8, 8))

#クラスタ別に色分け
color_codes = {0:"red",1:"blue",2:"green",3:"yellow"}
colors = [color_codes[x] for x in group_num]
plt.scatter(X,Y,color=colors)

for i,(x_name,y_name) in enumerate(zip(X,Y)):
    plt.annotate(df_skill.index[i],(x_name,y_name))
plt.show()
```

### K-meansの最適なクラスタ数推測（エルボー法）

```python
SSE = []
for i in range(1,11): 
    km = KMeans(n_clusters = i, n_init='auto')
    km.fit(df_skill)
    SSE.append(km.inertia_)

plt.plot(range(1,11),SSE,marker="o")
plt.xticks(range(1,11))
plt.xlabel("Number of clusters")
plt.ylabel("SSE")
plt.show()
```

### X-means

```python
#適当なデータの生成
random_data = np.concatenate((
    np.random.uniform(10, -10,  (50, 2)),
    np.random.uniform(20, -20,  (100, 2)) + [0, 100],
    np.random.uniform(30, -30,  (150, 2)) + [100, 0],
    np.random.uniform(40, -40,  (200, 2)) + [100, 100],
    ), axis=0)
```

```python
sns.scatterplot(data=pd.DataFrame(random_data),x=0,y=1)
```

```python
import pyclustering
from pyclustering.cluster.xmeans import xmeans
from pyclustering.cluster.center_initializer import kmeans_plusplus_initializer

xm_c = kmeans_plusplus_initializer(random_data, 2).initialize()
xm_i = xmeans(data=random_data, initial_centers=xm_c, ccore=False)
xm_i.process()

classes = len(xm_i._xmeans__centers)
predict = xm_i.predict(random_data)

clusters = xm_i.get_clusters()
pyclustering.utils.draw_clusters(random_data, clusters)
```

```python
# K-meansでもやってみる
df_sample = pd.DataFrame(random_data)

vec = KMeans(n_clusters = 5, n_init='auto')
group_num = vec.fit_predict(df_sample)

X = random_data[:,0]
Y = random_data[:,1]

sns.set(font="Hiragino Maru Gothic Pro",context="talk")
plt.figure(figsize=(8, 8))

#クラスタ別に色分け
color_codes = {0:"red",1:"blue",2:"green",3:"yellow",4:"black"}
colors = [color_codes[x] for x in group_num]
plt.scatter(X,Y,color=colors)
```

# ■アソシエーション分析（バスケット分析）

## 分析の実行

```python
#データの読み込み＆整形
df_sales = pd.read_csv("2-2_fruits_sales.csv",index_col="名前")
```

```python
#形式変換
df_sales = df_sales.fillna(False).replace("○",True)
```

```python
from mlxtend.frequent_patterns import apriori, association_rules, fpgrowth

#aprioriアルゴリズムでsupportが高い単品or組み合わせを選出
freq_items = apriori(df_sales, min_support=0.1, use_colnames=True)

#上で選ばれた組み合わせの中でliftが高い組み合わせを選出
freq_items_top = association_rules(freq_items, 
                                   metric = "confidence",
                                   min_threshold = 0.5,
                                   num_itemsets=freq_items.shape[0])

freq_items_top
```

## ネットワーク図の作成

```python
import networkx as nx

# 親ノードの抽出
ant = freq_items_top['antecedents'].values
ant = [tuple(x) for x in ant]

# 子ノードの抽出
con = freq_items_top['consequents'].values
con = [tuple(x) for x in con]

# 全ノードのリストアップ
both = list(set(ant + con))
```

```python
# 関係グラフの初期化
G = nx.DiGraph()

# ノードの追加
for n in both:
    G.add_node(n)

# エッジの追加
for i in range(len(freq_items_top)):
    item = freq_items_top.loc[i]
    ant = tuple(item['antecedents'])
    con = tuple(item['consequents'])
    G.add_edge(ant, con)
```

```python
# グラフの描画
pos = nx.spring_layout(G,seed=0,k=1.0)

plt.figure(figsize=(15, 10))

nx.draw_networkx_nodes(G, pos)
nx.draw_networkx_edges(G, pos)
nx.draw_networkx_labels(G, pos,
    horizontalalignment='left', 
    verticalalignment='center',
    font_family='Hiragino Maru Gothic Pro')

plt.axis('off')
plt.tight_layout()
plt.show()
```
